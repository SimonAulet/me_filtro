```python
from scipy import signal
import matplotlib.pyplot as plt
from math import pi
```

```python
import numpy as np
```

### Valores circuito

```python
R = 10e3
C_b = 560e-12
C_a = 22.3e-9
```

## Funciones de transferencia:

Para el pasa-altos: $\displaystyle H(\omega) = \frac{sRC}{1 + sRC}$

Y para el pasabajos: $\displaystyle \frac 1 {1 + sRC}$

## Ploteo pasaaltos y pasabajos

Se plotean en rojo el pasabajos y el azul el pasaaltos

```python
filtro = {'freq': np.logspace(np.log10(4e2), np.log10(4e6), num=10000)}

num = [R*C_a, 0]
den = [R*C_a, 1]
filtro['freq'], filtro['pasaaltos'], dummy = signal.bode((num, den), w=filtro['freq'])

num = [1]
den = [R*C_b, 1]
filtro['freq'], filtro['pasabajos'], dummy = signal.bode((num, den), w=filtro['freq'])

filtro['freq'] = np.array(filtro['freq']/(2*pi))#Convierto frecuencia a Hz
```

```python
ylim=-20

fig, ax = plt.subplots(figsize=(24, 8))
ax.set(ylim=ylim)
twin1 = ax.twinx()
twin1.set(ylim=ylim)
ax.semilogx(filtro['freq'], filtro['pasaaltos'], color='g')
twin1.semilogx(filtro['freq'], filtro['pasabajos'], color='r')
plt.grid(True)
plt.savefig('figs/'+'altos_bajos_superpuestos.png', format='png')
plt.savefig('figs/'+'altos_bajos_superpuestos.pdf', format='pdf')
plt.plot();
```

## Pasabanda
Como la señal es en todo momento afectada por los dos filtros; se suman las ganancias (negativas) de ambos filtros para formar el verdadero filtro (ideal) medido.

```python
filtro['pasabanda'] = filtro['pasaaltos'] + filtro['pasabajos']
```

```python
ylim=-20

fig, ax = plt.subplots(figsize=(24, 8))
ax.set(ylim=ylim)
#twin1 = ax.twinx()
#twin1.set(ylim=ylim)
ax.semilogx(filtro['freq'], filtro['pasabanda'], color='b')
#twin1.semilogx(filtro['freq'], filtro['pasabajos'], color='r')
plt.grid(True)
plt.savefig('figs/'+'pasabanda.png', format='png')
plt.savefig('figs/'+'pasabanda.pdf', format='pdf')
plt.plot();
```

## Superposicion
Superponemos los tres diagramas para ver la diferencia entre los filtros individuales y sumados:

```python
ylim=-3
xlim=(3e2, 10e4)

fig, ax = plt.subplots(figsize=(24, 8))
ax.set(xlim=xlim, ylim=ylim)
twin1 = ax.twinx()
twin1.set(xlim=xlim, ylim=ylim)
twin2 = ax.twinx()
twin2.set(xlim=xlim, ylim=ylim)

ax.semilogx(filtro['freq'], filtro['pasaaltos'], color='g')
twin1.semilogx(filtro['freq'], filtro['pasabajos'], color='r')
twin2.semilogx(filtro['freq'], filtro['pasabanda'], color='b')
plt.grid(True)

plt.savefig('figs/'+'todos_superpuestos.png', format='png')
plt.savefig('figs/'+'todos_superpuestos.pdf', format='pdf')

plt.plot();
```

Se ve cómo unicamente en el pico parece afectar la ganancia entre el pasabajos y el pasaaltos, el resto es prácticamente igual.
También se vé como la frecuencia de corte de $-3dB$ está en los $700\text{Hz}$ para las frecuencias bajas y $28k\text{Hz}$ para frecuencias altas, tal como dice la fórmula

# Comparacion con los datos reales

Vamos a buscarle la vuelta para extraer la ganancia para frecuencias buscadas. Al no ser un data frame no es tan directo pero calculo que se puede.<br>
Diossss como amo chat-gpt, con `numpy.searchsorted` se encuentra fácilmente. Vamos a poner una funcion para que imprima la frecuencia real y la ganancia del pasabanda a partir de una frecuencia deseada:

```python
def get_gain(freq, filtro):
    pos = np.searchsorted(filtro['freq'], freq)
    print('Frecuencia: {:.2f}Hz\nGanancia: {:.2f}dB'.format(filtro['freq'][pos], filtro['pasabanda'][pos]))
```

```python
get_gain(100, filtro)
```

```python
get_gain(400, filtro)
```
