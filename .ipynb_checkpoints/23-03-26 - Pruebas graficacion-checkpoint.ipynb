Vamos a probar graficar una convolucion

La fórmula es:

$y(t) = \displaystyle \int_1^{i+1} e^{-(\tau - 1)} d\tau$ para $y>0$

que debería quedar la función $\displaystyle 1 - e^{-t}$

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import *
```

```python
t = np.arange(0, 6, 0.01)
y_1 = 1 - e**(-t)
```

```python
#plt.figure(figsize=(120,120))

fig, ax = plt.subplots(figsize=(40,12))
fig.subplots_adjust(right=0.75)

twin1 = ax.twinx()

p1, = ax.plot(t, y_1, "C0", label="y_1")
p2, = twin1.plot(t, y_1, "C1", label="y_2")

ax.set(xlim=(0, 6), ylim=(0, 1.1), xlabel="t", ylabel="y")
twin1.set(xlim=(0, 6), ylim=(0, 1.1))

ax.legend(handles=[p1, p2])
plt.show()
```
