# Ejercicio 11
Grafique las siguientes señales en Python para $-1 < n < 5$

1. $x[n] = u[n]$ [La función escalón]
2. $x[n] = nu[n]$
3. $x[n] = 2^n u[n]$
4. $x[n] = sin(\pi n / 4)$
5. $x[n] = \delta[n] + \delta[n-1] + \delta[n-2]$
6. $x[n] = u[n] - u[n-3]$

```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
```

```python
heaviside = lambda t: np.heaviside(t, 1)
```

### 1. $x[n] = u[n]$

```python
t = np.arange(-10, 90, 1)
heaviside = lambda t: np.heaviside(t, 1)
```

```python
x_1 = list(map(heaviside, t))
x_1 = np.array(x_1)
```

```python
fig, ax =  plt.subplots(1, 1, figsize=(40, 12))
plot_1 = ax.stem(t, x_1)
ax.set(xlim=(-12, 100), ylim=(-1, 3));
```

### 2. $x[n] = nu[n]$

```python
n = np.arange(-8, 32, 1)

#vector np con funcion escalon aplicada
x_2u = list(map(heaviside, n))
x_2u = np.array(x_2u)

#x_2u
```

```python
#Estiro vector multiplicando por n

x_2 = x_2u * n
#x_2
```

```python
#a plotearrrr

fig, ax = plt.subplots(figsize=(40, 12))
plot2 = ax.stem(n, x_2)
ax.set(xlim=(-12, 48), ylim=(-8, 48));
```

## 3. $x[n] = 2^n u[n]$

```python
n_3 = np.arange(-6, 12, 1)
x_3 = np.ones(6 + 12)
x_3 = x_3 * 2
x_3 = x_3 ** n_3
x_3 = np.array(list(map(heaviside, n_3))) * x_3
```

```python
#a plotearrrr

fig, ax = plt.subplots(figsize=(40, 12))
plot3 = ax.stem(n_3, x_3)
ax.set(xlim=(-8, 18), ylim=(-8, 2e3));
```

## 4. $x[n] = sin(\pi n / 4)$

```python
n_4 = np.arange(-12, 36, 1)
x_4 = np.sin(np.pi * n_4 / 4)
```

```python
#a plotearrrr

fig, ax = plt.subplots(figsize=(40, 12))
plot4 = ax.stem(n_4, x_4)
ax.set(xlim=(-8, 32), ylim=(-1.5, 1.5));
```

## 5. $x[n] = \delta[n] + \delta[n-1] + \delta[n-2]$

```python
def dirac(n):
    if n == 0:
        return 1
    else:
        return 0
```

```python
n_5 = np.arange(-3, 6, 1)
x_5 = (np.array(list(map(dirac, n_5)))+np.array(list(map(dirac, n_5-1)))+np.array(list(map(dirac, n_5-2))))
```

```python
#a plotearrrr

fig, ax = plt.subplots(figsize=(40, 12))
plot5 = ax.stem(n_5, x_5)
ax.set(xlim=(-3, 6), ylim=(-0.5, 1.5));
```

## 6. $x[n] = u[n] - u[n-3]$

```python
n_6 = np.arange(-3, 6, 1)
x_6 = (np.array(list(map(heaviside, n_6)))-np.array(list(map(heaviside, n_6-3))))
```

```python
#a plotearrrr

fig, ax = plt.subplots(figsize=(40, 12))
plot6 = ax.stem(n_6, x_6)
ax.set(xlim=(-3, 6), ylim=(-0.5, 1.5));
```
